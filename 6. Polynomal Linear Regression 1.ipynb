```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("c:/Datasets/Salary_Experience.csv")
df.head()
```

```python
x = df.iloc[:, 0:1].values
x
```

```python
y = df.iloc[:, 1].values
y
```

```python
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary v/s Years of Experience")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 5)
```

```python
x1 = pf.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x1, y)
```

```python
lr.score(x1, y)
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x1))
```

```python
print("Coefficients :", lr.coef_)
```

```python
print("Intercept : ", lr.intercept_)
```

```python
input = [[5.5]]
input1 = pf.fit_transform(input)
result = lr.predict(input1)
print("Salary for 5.5 Yrs experience :", result)
```
