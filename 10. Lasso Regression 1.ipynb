```python
import pandas as pd
df = pd.read_csv("C:/Datasets/Advertising.csv")
df.head()
```

```python
names = df.columns
names
```

```python
df.rename({"Unnamed: 0" : "a"}, inplace = True, axis = 1)
df.drop(["a"], inplace = True, axis = 1)
```

```python
x = df.drop(["sales"], axis = 1)
x
```

```python
y = df["sales"].values
y
```

```python
names = x.columns
names
```

```python
rng = range(len(names))
rng
```

```python
from sklearn.linear_model import Lasso
ls = Lasso(alpha = 0.1)
model = ls.fit(x, y)
cf = ls.coef_
cf
```

```python
import matplotlib.pyplot as plt
plt.plot(rng, cf)
plt.xticks(rng, names, rotation = 60)
plt.ylabel("Coefficients")
plt.show()
```

```python
ls.score(x, y)
```

```python
newdata = [[150, 41, 60]]
sales = ls.predict(newdata)
sales
```
