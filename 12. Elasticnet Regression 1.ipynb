```python
import pandas as pd
df = pd.read_csv("C:/Datasets/diabetes.csv")
df.head()
```

```python
x = df.drop(["Outcome"], axis = 1)
x
```

```python
y = df["Outcome"].values
y
```

```python
names = x.columns
names
```

```python
rng = range(len(names))
rng
```

```python
from sklearn.linear_model import ElasticNet
es = ElasticNet(l1_ratio = 0.5)
```

```python
model = es.fit(x, y)
```

```python
es_coef = model.coef_
es.coef_
```

```python
import numpy as np
n = np.argmax(es_coef)
n
```

```python
print("The most influencing column :", names[n])
```

```python
import matplotlib.pyplot as plt
plt.plot(rng, es_coef)
plt.xticks(rng, names, rotation = 60)
plt.ylabel("Coefficients")
plt.show()
```
